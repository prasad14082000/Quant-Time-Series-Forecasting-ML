```python
import yfinance as yf
```

```python
symbol = 'NVDA'
start = '2015-01-01'
end = '2024-01-01'
```

```python
dataset = yf.download(symbol,start,end)
dataset.head()
```

```python
df = dataset.drop(['High','Low','Volume'], axis = 1)
df.head()
```

```python
a = np.array(df)
a
```

```python
print("a = ", a)
print("a[1] = ", a[1])
print("a[1][1] = ", a[1][1])
print("a[0][-1] = ", a[0][-1])
```

```python
column = [];
for row in a:
    column.append(row[1])

print("Second column = ", column)
```

```python
a = np.arange(4)
print('a = ', a)

b = np.arange(12).reshape(2,6)
print('b = ', b)
```

```python
A = np.column_stack((dataset['High'], dataset['Low']))
B = np.column_stack((dataset['Open'], dataset['Close']))

print(A)
print(B)
```

```python
c = A + B
print("Addition Matrix: \n", c)
```

```python
d = A - B
print("Substraction Matrix: \n", d)
```

```python
e = A.dot(B.T)
print("Multiplication of two matrices: \n", e)
```
